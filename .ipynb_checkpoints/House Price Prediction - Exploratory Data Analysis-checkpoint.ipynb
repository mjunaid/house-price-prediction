```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

##Display all columns of the dataframe

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')

print(dataset.shape)
```

```python
# Let's briefly check the data
dataset.head()
```

```python
# What variables have missing values in the data?
variables_with_na=[variables for variables in dataset.columns if dataset[variables].isnull().sum()>=1]
print(variables_with_na)
```

```python
# Quickcheck to see how many records 'Alley' has values missing for
print(dataset['Alley'].isnull().sum())
```

```python
# What % of the dataset is missing values for each variable with na?
for variable in variables_with_na:
    print(variable, np.round(dataset[variable].isnull().mean()*100,2), '% missing values')
```

```python
# We can see from above that Alley, PoolQC and MiscFeature data are top 3 variables with missing data
```

```python
# Let's see the impact of these missing values on the sales price

for variable in variables_with_na:
    data = dataset.copy()

    # Classify the data as null and not null
    data[variable] = np.where(data[variable].isnull(),'Null','Not Null')

    # group (aggregate) the data by variable name and take the median of the sale price

    data.groupby(variable)['SalePrice'].median().plot.bar()
    plt.title(variable)
    plt.show()
```

```python
# From the results we can see there is some relationship between the missing values and Sales Price. 
# For example, GarageCond with Null values have a lower median sales price. 
# This makes sense because one hypothesis could be that where garange condition is null, there may not be a garage, 
# hence the property is being sold at a lower rate.
```
